```python
from pypfopt import expected_returns
from pypfopt import risk_models
from pypfopt.efficient_frontier import EfficientFrontier
from pypfopt.discrete_allocation import DiscreteAllocation, get_latest_prices
from pandas_datareader import data
```

```python
assets = ['BTC-USD', 'ETH-USD', 'SOL-USD', 'LUNA1-USD', 'ONE1-USD']
```

```python
df = data.DataReader(assets, "yahoo", "2021-01-01")['Adj Close']
df
```

```python
# expected simple annualized return since 2021-01-01
mu = expected_returns.mean_historical_return(df)
mu
```

```python
# sample covariance matrix
S = risk_models.sample_cov(df)
S
```

```python
ef = EfficientFrontier(mu, S, weight_bounds=(0,1))
ef
```

```python
ef.max_sharpe() # performing optmization
# suggest weights of each asset in the portfolio
# based on maximizing Sharpe ratio
```

```python
_ = ef.portfolio_performance(verbose=True)
```
